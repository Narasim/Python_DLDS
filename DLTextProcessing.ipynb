# One-hot Encoding

## Word Level One-hot Encoding

```python
import numpy as np
samples = ['The cat sat on the mat', 'The dog ate my homework.']
tokenIndex = {}

for sample in samples:
    for word in sample.split():
        if word not in tokenIndex:
            tokenIndex[word] = len(tokenIndex)+1
maxLength = 10
results = np.zeros(shape = (len(samples), maxLength, max(tokenIndex.values())+1))
for i, sample in enumerate(samples):
    for j, word in list(enumerate(sample.split())):
        index = tokenIndex.get(word)
        results[i, j, index] = 1

results
```

## One-Hot Encoding for Categorical Data Using In-Built Function

```python
import tensorflow as tf
categories = [0, 1, 2, 1, 3]
numCategories = 4
oneHot = tf.keras.utils.to_categorical(categories, num_classes = numCategories)
print(oneHot)
```

## Wor-level One-Hot Encoding for Text Data Using BuiltIn Functions

```python
from tensorflow.keras.preprocessing.text import Tokenizer
samples = ['The cat sat on the mat', 'The dog ate my homework']
tokenizer = Tokenizer(num_words = 10)
tokenizer.fit_on_texts(samples)
print(tokenizer.word_index)
print(tokenizer.word_index.keys())
sequences = tokenizer.texts_to_sequences(samples)
print(sequences)
oneHotResults = tokenizer.texts_to_matrix(samples, mode = 'binary')
print(oneHotResults)
```

# Word Embeddings

```python
from keras.layers import Embedding
from keras.datasets import imdb
from keras import preprocessing
from keras.models import Sequential
from keras.layers import Flatten, Dense
```

```python
maxFeatures = 10000
maxLen = 20
(xTrain, yTrain), (xTest, yTest) = imdb.load_data(num_words = maxFeatures)

xTrain = preprocessing.sequence.pad_sequences(xTrain, maxlen = maxLen)
xTest = preprocessing.sequence.pad_sequences(xTest, maxlen = maxLen)


model = Sequential()
model.add(Embedding(maxFeatures, 8, input_length = maxLen))
model.add(Flatten())
model.add(Dense(1, activation = 'sigmoid'))
model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])


model.summary()

results = model.fit(xTrain, yTrain,
                   epochs = 10,
                   batch_size = 32,
                   validation_split = 0.2)
```

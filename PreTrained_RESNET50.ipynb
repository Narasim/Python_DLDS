```python
from tensorflow.keras.applications import resnet50
```

```python
# prompt: import load_img, img_to_array, decode_predictions and tensorflow

from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.resnet50 import decode_predictions
import tensorflow as tf
import numpy as np
```

```python
imgPath = "/content/images.jpeg"
img = load_img(imgPath, target_size = (224, 224))
img_np = img_to_array(img)
img_np = np.expand_dims(img_np, axis = 0)
```

```python
model = resnet50.ResNet50(weights = 'imagenet')
X = resnet50.preprocess_input(img_np.copy())
y = model.predict(X)
predict_labels = decode_predictions(y)
print(predict_labels)
```

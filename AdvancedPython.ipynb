```python
import matplotlib.pyplot as plt
items  = ['name', 'fame', None, 'game', 'same', 'kagg']
print(items)
```

```python
def word_magic(word):
    magic = 0
    for letter in word:
        magic += ord(letter) - ord('a')
    return magic
magic_nums = []        
for word in items:
    if(word != None):
        magic_nums.append(word_magic(word))
        
plt.plot(magic_nums)
plt.title("Magic Numbers")
plt.show()
```

## Use enumerate() Instead of Looping Over an Index

```python
for i in range(len(items)):
    print(i, items[i])
    
for i in items:
    print(None, i)
    
for i, item in enumerate(items):
    print(i, item)
    
for i, _ in enumerate(items):
    print(i, None)
```

## Don’t Iterate Over dict.keys() When You Want dict.items()

```python
my_dict = {letter:ord(letter) for letter in 'Narasimhulu'}
print(my_dict)
for i in iter(my_dict.keys()):
    print(i)
```
